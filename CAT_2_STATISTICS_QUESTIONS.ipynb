```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, chi2
```

```python
# QUESTION 1
xbar = 82
sigma = 4
n = 36
z = 1.96

se = sigma / np.sqrt(n)
lower = xbar - z * se
upper = xbar + z * se

x = np.linspace(78, 86, 400)
y = norm.pdf(x, xbar, se)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= lower) & (x <= upper))
plt.axvline(lower)
plt.axvline(upper)
plt.title("Q1: 95% Confidence Interval")
plt.xlabel("Transaction Time (s)")
plt.show()

print("95% CI:", lower, upper)
```

```python
#Q2
mu = 128.5
sigma = 1.5

x = np.linspace(123, 134, 400)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 130))
plt.axvline(130)
plt.title("Q2: P(X > 130g)")
plt.xlabel("Mass (g)")
plt.show()

print("Probability:", 1 - norm.cdf(130, mu, sigma))
```

```python
# Q3
mu = 180
sigma = 20
n = 4

se = sigma / np.sqrt(n)

x = np.linspace(150, 210, 400)
y = norm.pdf(x, mu, se)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 175) & (x <= 190))
plt.axvline(175)
plt.axvline(190)
plt.title("Q3: Sample Mean Probability")
plt.xlabel("Sample Mean (ms)")
plt.show()

print("Probability:", norm.cdf(190, mu, se) - norm.cdf(175, mu, se))
```

```python
# Q4
mu0 = 200
xbar = 206
sigma = 24
n = 64

z_stat = (xbar - mu0) / (sigma / np.sqrt(n))

x = np.linspace(-4, 4, 400)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -1.96))
plt.fill_between(x, y, where=(x >= 1.96))
plt.axvline(z_stat)
plt.title("Q4: Two-Tailed Z-Test")
plt.xlabel("Z")
plt.show()

print("Z statistic:", z_stat)
```

```python
# Q5
mu = 120
sigma = 30
n = 64

se = sigma / np.sqrt(n)

x = np.linspace(100, 140, 400)
y = norm.pdf(x, mu, se)

plt.plot(x, y)
plt.fill_between(x, y, where=(x >= 115) & (x <= 130))
plt.title("Q5: CLT Sampling Distribution")
plt.xlabel("Sample Mean")
plt.show()

print("Probability:", norm.cdf(130, mu, se) - norm.cdf(115, mu, se))
```
