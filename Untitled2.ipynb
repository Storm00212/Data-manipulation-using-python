```python
# A simulation of valentine's day using math functions
#deps
import numpy as np
import matplotlib.pyplot as plt

# Parameter
k = 92.75

# Proper domain restriction
x = np.linspace(-np.sqrt(3), np.sqrt(3), 4000)

# Fix power issue using absolute value
base = np.abs(x)**(2/3)

# Ensure sqrt stays real
inside = np.sqrt(np.maximum(0, 3 - x**2))

# Top and bottom curves
y_top = base + 0.9*np.sin(k*x)*inside
y_bottom = -base + 0.9*np.sin(k*x)*inside

# Plot
plt.figure(figsize=(6,6))
plt.style.use("dark_background")

# Vertical stripe effect
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y_bottom[i], y_top[i]], color="red", linewidth=0.6)

# Axes
plt.axhline(0, color="white", linewidth=1)
plt.axvline(0, color="white", linewidth=1)

plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.title("Happy Valentines!", color="lightcoral", fontsize=18)

plt.show()
```
