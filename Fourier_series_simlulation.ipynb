```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Create x values
x = np.linspace(-np.pi, np.pi, 1000)

# Number of Fourier terms
N = 10   # Increase this to improve approximation
```

```python
# Fourier series approximation
square_wave = np.zeros(len(x))

for n in range(1, N+1):
    harmonic = 2*n - 1   # Odd numbers: 1,3,5,...
    square_wave += (4/np.pi) * (1/harmonic) * np.sin(harmonic * x)
```

```python

# Plotting
plt.figure(figsize=(10,6))
plt.plot(x, square_wave, label=f'Fourier Approximation (N={N})')
plt.title("Fourier Series Approximation of a Square Wave")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
